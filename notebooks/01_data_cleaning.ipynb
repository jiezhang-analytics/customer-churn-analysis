# 01 Data Cleaning and Preprocessing

## Project Context
This notebook prepares the Netflix customer churn dataset for downstream analysis and modeling.

## Objectives
- Load and inspect the raw dataset
- Check structure, data types, missing values, and duplicates
- Clean unnecessary fields
- Handle basic data quality issues
- Save a cleaned dataset for EDA and modeling

```python
import pandas as pd
import numpy as np
```

## 1. Load the Dataset

```python
data_path = "../data/raw/netflix_customer_churn.csv"
df_raw = pd.read_csv(data_path)

df_raw.head()
```

## 2. Initial Data Inspection

```python
print("Dataset shape:", df_raw.shape)
print("\nColumn names:")
print(df_raw.columns)
```

```python
df_raw.info()
```

```python
df_raw.describe(include="all").T
```

## 3. Check Missing Values and Duplicates

```python
missing_values = df_raw.isnull().sum()
missing_values
```

```python
duplicate_rows = df_raw.duplicated().sum()
print("Number of duplicate rows:", duplicate_rows)
```

## 4. Check Target Variable Distribution

```python
target_distribution = df_raw["churned"].value_counts()
target_ratio = df_raw["churned"].value_counts(normalize=True)

print("Target counts:")
print(target_distribution)

print("\nTarget ratio:")
print(target_ratio)
```

## 5. Create a Working Copy
We keep the original dataset unchanged and perform cleaning on a copied version.

```python
df_clean = df_raw.copy()
```

## 6. Remove Non-Predictive Identifier
`customer_id` is a unique identifier and does not provide predictive value for analysis or modeling.

```python
df_clean = df_clean.drop(columns=["customer_id"])
df_clean.head()
```

## 7. Validate Data Quality
This step checks for:
- missing values
- invalid ranges
- basic logical consistency

```python
print("Missing values after dropping customer_id:")
print(df_clean.isnull().sum())
```

```python
print("Unique values for categorical columns:\n")

categorical_columns = df_clean.select_dtypes(include="object").columns

for column in categorical_columns:
    print(f"{column}: {df_clean[column].unique()}")
```

## 8. Handle Potential Data Issues
The original project clipped `avg_watch_time_per_day` at 24 hours.
This is a reasonable rule because daily watch time cannot logically exceed 24 hours.


```python
df_clean["avg_watch_time_per_day"] = df_clean["avg_watch_time_per_day"].clip(upper=24)
```

```python
df_clean["avg_watch_time_per_day"].describe()
```

## 9. Optional Missing Value Treatment
The original dataset does not contain missing values, but we include a safe step for reproducibility.
For categorical columns, mode imputation is applied only if missing values exist.

```python
if df_clean.isnull().sum().sum() > 0:
    df_clean = df_clean.fillna(df_clean.mode().iloc[0])

print("Remaining missing values:", df_clean.isnull().sum().sum())
```

## 10. Final Review of Cleaned Dataset

```python
print("Cleaned dataset shape:", df_clean.shape)
df_clean.head()
```

```python
df_clean.info()
```

## 11. Save Cleaned Dataset
This cleaned dataset will be used in later notebooks for EDA, modeling, and business analysis.

```python
output_path = "../data/processed/netflix_churn_cleaned.csv"
df_clean.to_csv(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```
